```python
import pandas as pd
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

df = pd.read_csv('results.csv')

#GRAFICO 1

plt1.title('PARTIDOS INTERNACIONALES', fontsize=22)
plt1.xlabel('AÑO', fontsize=12)
plt1.ylabel('CANTIDAD PARTIDOS', fontsize=12)

df['date_time'] = pd.to_datetime(df.date)

df.date_time.dt.year.value_counts().sort_index().plot()
plt1.show()

#GRAFICO 2

df2 = pd.read_csv('results.csv')

plt2.title('PARTIDOS INTERNACIONALES', fontsize=22)
plt2.xlabel('AÑO', fontsize=12)
plt2.ylabel('CANTIDAD PARTIDOS', fontsize=12)

df2['date_time'] = pd.to_datetime(df2.date)

df2.date_time.dt.year.value_counts().sort_index().bar()
plt2.show()

#¿Qué conclusión o conclusiones sacaría del gráfico obtenido?
# A medida que pasa el tiempo, la cantidad de partidos internacionales son mayores, a exepcion del 2020 al haberse frenado la actividad por la pandemia.

#¿Cuáles serían las variables que considera que deberían tenerse en cuenta pero no las puede apreciar porque el csv no se las provee? ¿Porqué?
# Para este grafico las variables que provee el dataset alcanza.

#¿Podría relacionar los datos de las dos columnas elegidas para especular y proponer una conclusión? En caso de que no, ¿Con qué columnas cree que lo podría hacer? ¿Porqué?
# Podrias relacionar las columnas para ver los años mundialistas, ya que esos años son los que mas cantidad de partidos internacionales tiene.




```
