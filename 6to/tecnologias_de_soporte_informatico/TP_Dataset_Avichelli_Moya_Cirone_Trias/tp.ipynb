```python
import pandas as pd

df = pd.read_csv("respuestas.csv")

grafico1 = df['En que plataforma/s jugas?']
grafico1.value_counts().plot(kind="bar", title="Plataformas en las que se juega", legend="true")

# DESCRIPCION DEL GRÁFICO:
# A raíz de este gráfico podemos deducir que la plataforma mas utilizada por los "gamers" es la PC, mientras que la segunda más utilizada vendría a ser la Playstation,
# seguida de las Nintendo y el Celular, estas últimas siendo las menos utilizadas.
```

```python
import pandas as pd

df = pd.read_csv("respuestas.csv")

grafico1 = df['Cuantas horas le invertis generalmente cuando te sentas a jugar?']
grafico1.value_counts().plot(kind="pie", legend="true")

# DESCRIPCION DEL GRÁFICO:
# El gráfico representa en modo de torta la cantidad de horas diarias las cuales una persona dedica a jugar videojuegos. Según los resultados obtenidos,
# la mayoria de la gente dedica 2 a 3 horas de su tiempo a jugar videojuegos, mientras que otro cuarto de las respuestas muestran de 1-2hs y de 3-4hs, 
# y con una minoria de la gente la cual juega más de 4hs diarias.
```

```python
import pandas as pd

df = pd.read_csv("respuestas.csv")

grafico1 = df['Genero/s de videojuegos preferido/s?']
grafico1.value_counts().plot(kind="bar", legend="true")

# DESCRIPCION DEL GRÁFICO:
# En el siguiente gráfico se puede observar los distintos géneros de videojuegos, ordenados de más jugados a menos jugados. 
# El género de videojuegos mas jugado es el de "Shooter/Accion" con una media de 5 personas, mientras que en el segundo lugar, con una media de 3 personas, están los juegos de "Estrategia", 
# seguido de los de "Rol" y "Aventura" con 2 de media, y por último estan los juegos de "Deporte", "Automovilismo" y "Simulación" con 1 de media de jugadores.
```

```python
import pandas as pd

df = pd.read_csv("respuestas.csv")
grafico1 = df['A que edad comenzaste a jugar videojuegos?']

grafico1.value_counts().plot(kind="bar", legend="true")

# DESCRIPCION DEL GRÁFICO:
# En el siguiente gráfico podemos observar a que edad la gente comenzo a jugar videojuegos, teniendo como rango de edad desde los 6 años a los 10 años.
# La mayoria comenzó a jugar a los 6 años de edad, con una media de 5 personas. Luego en segundo lugar estan los 8 años de edad con una media de 4 personas. 
# En tercer lugar y cuarto lugar tenemos a los que comenzaron a los 4 y 7 años con una media de 2. Y por ultimo tenemos a la gente que comenzó a jugar videojuegos a los 10 y 9 años de edad.
```
